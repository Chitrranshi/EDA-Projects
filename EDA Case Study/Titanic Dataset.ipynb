```python

def Age(x):
    'divide the age into four categories'
    if x < 18:
        return "child"
    elif 18 <= x < 25:
        return "Teenager"
    elif 25 <= x < 45:
        return "Young"
    elif 45 <= x < 100:
        return "Senior"


df['age_slot'] = df.Age.apply(lambda x: Age(x))
```
